```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import re # Regular expressions
```

```python
paragraph ="""
By Statistics, we mean methods specially adopted to the 
elucidation of quantitative data affected to a marked extent by 
multiplicity of causes”.
Yule and Kendal
It is interesting to see what Thomas Davenport means by Business 
Analytics and note the similarities and dissimilarities between 
the two.
“Business Analytics (BA) can be defined as the broad use of 
data and quantitative analysis for decision making within organisation
"""
```

```python
sentences= sent_tokenize(paragraph)
words=word_tokenize(paragraph)
stemmer=PorterStemmer()
lemmatizer=WordNetLemmatizer()
corpus= []
```

```python
# stemming
for i in range(len(sentences)):
    review=re.sub('[^a-zA-Z]', ' ',sentences[i])
    review=review.lower()
    review=review.split()
    review=[stemmer.stem(word) for word in review if  not word in set(stopwords.words("english"))  ]
    review=" ".join(review)
    corpus.append(review)
```

```python
import nltk
from nltk import tokenize
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
import pandas as pd
import numpy
import re
stemmer=PorterStemmer()
Lemmitizer=WordNetLemmatizer()
```

```python
messages=pd.read_csv("SMSSpamCollection", sep='\t', names=["Lables", "Messages"])
```

```python
messages
```

```python
messages.head()
```

```python
corpus=[]
```

```python
#  Stemming
for i in range(0,len(messages)):
    review=re.sub('[^a-zA-Z]', ' ',messages['Messages'][i])
    review=review.lower()
    review=review.split()
    review=[stemmer.stem(word) for word in review if  not word in set(stopwords.words("english"))]
    review=" ".join(review)
    corpus.append(review)
```

```python
corpus # Lemmitization
```

```python
corpus # Stemming
```

```python
# Bog of Words

import sklearn
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=2500, binary=True)
X=cv.fit_transform(corpus).toarray()
```

```python
X
```

```python
y=pd.get_dummies(messages['Lables'])
```

```python
y.head()
```

```python
y=y.iloc[:,1].values
```

```python
y
```

```python
#  Train Test Split

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y, test_size=0.20, random_state=1)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred= spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_x=confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(y_test, y_pred)
```
